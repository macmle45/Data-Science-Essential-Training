```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

### Grouping data by column index

```python
address_path = "../../Data/mtcars.csv"
cars = pd.read_csv(address_path)

cars.columns = ['car_name', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs',
       'am', 'gear', 'carb']

cars.head()
```

```python
gr = cars.groupby(cars['am'])
gr.mean()
```
